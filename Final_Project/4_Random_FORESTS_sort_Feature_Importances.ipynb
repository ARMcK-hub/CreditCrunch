```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
default= pd.read_csv('Credit_Data_Original.csv')
default.head()
```

```python
new_default=default.drop("OBS#", axis=1)
new_default
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
default.data = new_default.drop("DEFAULT", axis=1)
default.target = new_default["DEFAULT"]
```

```python
# Create a random forest classifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(default.data, default.target)
rf.score(default.data, default.target)
```

```python
 # Random Forests in sklearn will automatically calculate feature importance
importances = rf.feature_importances_
importances
```

```python
# We can sort the features by their importance
sorted(zip(rf.feature_importances_, new_default), reverse=True)
```

```python
## MAYBE put in an ASSENDING dataFrame to show Most Important Columns to least when determining DEFAULT Outcomes.
FEATURE_DF= pd.DataFrame({"Prediction Importance Feature":default.data, "Importance Rating":importances})
FEATURE_DF
# To sort from highest to lowest, ascending=False must be passed in & resest index
new_feature_df= FEATURE_DF.sort_values("Importance Rating", ascending=False).reset_index(drop=True)
new_feature_df

# I don't know how the names have commas. The only way the dataFrame will read in and correspond to the importance rating
# has name and comma. 
```
